```python
def cal_train_val_test(num):
    if .7*num + .2*num + .1*num == num:
        train = round(.7 * num, 0)
        val = round(.2 * num, 0)
        test = round(.1 * num, 0)
    else:
        train = round(.7 * num, 0)
        val = round(.2 * num, 0)+1
        test = round(.1 * num, 0)+1
    return int(train), int(val), int(test)
```

```python
x,y,z = cal_train_val_test(500)
print(x,y,z)
```
